```python
f = open("dictionary.txt", "r")
dictionary = [i[:-1] for i in f]
```

```python
size = 6
d = [i for i in dictionary if len(i) == size]
```

```python
def head ( string ):
    return string[1:3]
def tail ( string ):
    return string[-3:-1]
def first (dictionary):
    result = {}
    for i in dictionary:
        sub = head(i)
        if result.get(sub):
            result[sub].append(i)
        else:
            result[sub] = [i]
    return result

def last (dictionary):
    result = {}
    for i in dictionary:
        sub = tail(i)
        if result.get(sub):
            result[sub].append(i)
        else:
            result[sub] = [i]
    return result
            
```

```python
def get(x, F):
    if F.get(x):
        return len(F[x])
    return 0
```

```python
def findM (words):
    posibilities = {} 
    for word in words:
        try:
            posibilities[word] = HEAD[tail(word)]
        except:
            posibilities[word] = []

    maxi = [i for i in words if (len(posibilities[i]) == len(posibilities[max(posibilities, key = lambda k: len(posibilities[k]))])) ]
    return maxi
```
